```
#Numero Primo - Refatorado
import math

def eh_primo(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    # Verificamos apenas até a raiz quadrada de n
    limite = int(math.sqrt(n)) + 1
    for i in range(3, limite, 2):
        if n % i == 0:
            return False
    return True
```

```
# Testando a função eh_primo
test_cases = [2, 3, 4, 5, 10, 13, 17, 20]
for num in test_cases:
    print(f'{num} é primo? {eh_primo(num)}')
```

```
import requests
import json

def consultar_cep(cep):
    """
    Consulta informações de um CEP utilizando a API do ViaCEP.
    Args:
        cep (str): O número do CEP a ser consultado (apenas dígitos).
    Returns:
        dict: Um dicionário com os dados do endereço, ou None se houver erro.
    """
    # A API do ViaCEP aceita CEPs com ou sem hífen, mas é mais comum enviar apenas dígitos.
    # Limpa o CEP, removendo quaisquer caracteres não numéricos
    cep = ''.join(filter(str.isdigit, cep))

    if len(cep) != 8:
        print("Erro: O CEP deve conter 8 dígitos.")
        return None

    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Levanta um erro HTTP para status ruins (4xx ou 5xx)

        data = response.json()

        if 'erro' in data and data['erro'] == True:
            print(f"CEP {cep} não encontrado.")
            return None
        else:
            return data

    except requests.exceptions.HTTPError as http_err:
        print(f"Erro HTTP: {http_err} - Verifique o CEP ou a conexão.")
    except requests.exceptions.ConnectionError as conn_err:
        print(f"Erro de Conexão: {conn_err} - Verifique sua conexão com a internet.")
    except requests.exceptions.Timeout as timeout_err:
        print(f"Erro de Timeout: {timeout_err} - A requisição demorou muito.")
    except requests.exceptions.RequestException as req_err:
        print(f"Erro na requisição: {req_err}")
    except json.JSONDecodeError:
        print("Erro ao decodificar a resposta JSON.")
    return None

# --- Exemplo de Uso ---

# Exemplo 1: CEP válido
cep_valido = "01001-000" # Pode ser "01001000" também
endereco = consultar_cep(cep_valido)

if endereco:
    print(f"\n--- Informações para o CEP {cep_valido} ---")
    print(f"Logradouro: {endereco.get('logradouro')}")
    print(f"Bairro: {endereco.get('bairro')}")
    print(f"Cidade: {endereco.get('localidade')}")
    print(f"Estado: {endereco.get('uf')}")
    print(f"DDD: {endereco.get('ddd')}")

# Exemplo 2: CEP não encontrado
cep_nao_existente = "99999999"
endereco_nao_existente = consultar_cep(cep_nao_existente)

# Exemplo 3: CEP inválido (formato)
cep_invalido = "123"
endereco_invalido = consultar_cep(cep_invalido)
```

```
# Função para adição
def somar(x, y):
    return x + y

# Função para subtração
def subtrair(x, y):
    return x - y

# Função para multiplicação
def multiplicar(x, y):
    return x * y

# Função para divisão
def dividir(x, y):
    if y == 0:
        return "Erro: Não é possível dividir por zero!"
    else:
        return x / y

# Exemplo de uso da calculadora
print("Selecione a operação:")
print("1. Soma")
print("2. Subtração")
print("3. Multiplicação")
print("4. Divisão")

while True:
    # Pega a escolha do usuário
    escolha = input("Digite sua escolha (1/2/3/4): ")

    # Verifica se a escolha é válida
    if escolha in ('1', '2', '3', '4'):
        try:
            num1 = float(input("Digite o primeiro número: "))
            num2 = float(input("Digite o segundo número: "))
        except ValueError:
            print("Entrada inválida. Por favor, digite números.")
            continue

        if escolha == '1':
            print(f"{num1} + {num2} = {somar(num1, num2)}")

        elif escolha == '2':
            print(f"{num1} - {num2} = {subtrair(num1, num2)}")

        elif escolha == '3':
            print(f"{num1} * {num2} = {multiplicar(num1, num2)}")

        elif escolha == '4':
            resultado = dividir(num1, num2)
            print(f"{num1} / {num2} = {resultado}")

        # Pergunta ao usuário se ele quer continuar
        proxima_operacao = input("Deseja fazer outra operação? (sim/não): ")
        if proxima_operacao.lower() == "não":
            break
    else:
        print("Escolha inválida. Por favor, digite 1, 2, 3 ou 4.")
```

```
import random

def jogar_forca():
    # Lista de palavras secretas para o jogo
    palavras = ["PYTHON", "PROGRAMACAO", "COLAB", "INTELIGENCIA", "ARTIFICIAL", "ALGORITMO"]

    # Escolhe uma palavra aleatoriamente e a converte para maiúsculas para facilitar a comparação
    palavra_secreta = random.choice(palavras).upper()

    # Inicializa a lista de letras acertadas com underscores para cada letra da palavra secreta
    # Ex: se a palavra for "PYTHON", letras_acertadas será ['_', '_', '_', '_', '_', '_']
    letras_acertadas = ["_" for letra in palavra_secreta]

    # Lista para armazenar todas as letras que o usuário já tentou
    letras_tentadas = []

    # Número máximo de erros permitidos antes que o jogo termine
    tentativas_maximas = 6

    # Contador de erros atuais
    erros = 0

    print("Bem-vindo ao Jogo da Forca!")
    print(f"A palavra tem {len(palavra_secreta)} letras.")

    # Função refatorada para exibir o estado atual do jogo
    # A "refatoração via IA" aqui significa tornar a exibição mais modular, clara e formatada.
    def exibir_estado_jogo():
        # Exibe a palavra com as letras acertadas e os underscores para as letras não descobertas
        # O ' '.join(letras_acertadas) transforma a lista em uma string com espaços entre as letras/underscores
        print(f"\nPalavra: {' '.join(letras_acertadas)}")

        # Exibe as letras que já foram tentadas, ordenadas para melhor visualização
        print(f"Letras tentadas: {', '.join(sorted(letras_tentadas))}")

        # Mostra o número de erros atuais e o total de erros permitidos
        print(f"Erros: {erros}/{tentativas_maximas}")

        # Calcula e mostra o número de tentativas restantes
        print(f"Tentativas restantes: {tentativas_maximas - erros}")

        # Desenho simples da forca com base nos erros (pode ser expandido para um desenho mais detalhado)
        print("  _____")
        print("  |   |")
        if erros > 0:
            print("  O   |")
        else:
            print("      |")
        if erros == 2:
            print("  |   |")
        elif erros == 3:
            print(" /|   |")
        elif erros >= 4:
            print(" /|\  |")
        else:
            print("      |")
        if erros == 5:
            print(" /    |")
        elif erros >= 6:
            print(" / \  |")
        else:
            print("      |")
        print("      |")
        print("__|___")
        print("-" * 20)


    # Loop principal do jogo, continua enquanto o jogo não termina
    while True:
        exibir_estado_jogo() # Chama a função para mostrar o estado atual do jogo

        # Condição de vitória: se não houver mais underscores na palavra, o jogador venceu
        if "_" not in letras_acertadas:
            print(f"Parabéns! Você adivinhou a palavra: {palavra_secreta}")
            break # Sai do loop do jogo

        # Condição de derrota: se o número de erros atingir o máximo, o jogador perdeu
        if erros == tentativas_maximas:
            print(f"Fim de jogo! Você excedeu o número de erros.")
            print(f"A palavra secreta era: {palavra_secreta}")
            break # Sai do loop do jogo

        # Solicita ao usuário que digite uma letra e a converte para maiúscula
        tentativa = input("Digite uma letra: ").upper()

        # Validação da entrada do usuário
        # 1. Verifica se foi digitado apenas um caractere
        # 2. Verifica se o caractere é uma letra do alfabeto
        if len(tentativa) != 1 or not tentativa.isalpha():
            print("Entrada inválida. Por favor, digite apenas uma letra.")
            continue # Volta para o início do loop sem contar como tentativa

        # Verifica se a letra já foi tentada anteriormente
        if tentativa in letras_tentadas:
            print(f"Você já tentou a letra '{tentativa}'. Tente outra.")
            continue # Volta para o início do loop sem contar como tentativa

        # Adiciona a letra tentada à lista de letras já tentadas
        letras_tentadas.append(tentativa)

        # Verifica se a letra tentada está na palavra secreta
        if tentativa in palavra_secreta:
            print(f"Boa! A letra '{tentativa}' está na palavra.")
            # Percorre a palavra secreta para atualizar as letras acertadas
            for i, letra in enumerate(palavra_secreta):
                if letra == tentativa:
                    letras_acertadas[i] = tentativa # Substitui o underscore pela letra correta
        else:
            print(f"Que pena! A letra '{tentativa}' não está na palavra.")
            erros += 1 # Incrementa o contador de erros

# Chamada para iniciar o jogo quando o script é executado
jogar_forca()
```
