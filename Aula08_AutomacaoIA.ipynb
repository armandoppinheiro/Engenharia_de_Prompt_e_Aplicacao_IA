```python
'''
Crie um script Python simples para verificar a temperatura de um sistema.
Se for maior que 30, exiba um alerta de superaquecimento.
'''

print("Iniciando a verificação de temperatura do sistema...") # Comentário durante a execução

try:
    temperature_input = input("Por favor, digite a temperatura atual do sistema em °C: ")
    temperature = int(temperature_input)
    print("Temperatura atual definida como: " + str(temperature) + "°C") # Comentário e concatenação

    if temperature > 30:
        print("A temperatura está acima do limite de 30°C.") # Comentário durante a execução
        # Usando concatenação de strings para o alerta
        print("ALERTA: Temperatura do sistema (" + str(temperature) + "°C) está muito alta! Risco de superaquecimento!")
    else:
        print("A temperatura está dentro dos limites aceitáveis (até 30°C).") # Comentário durante a execução
        # Usando concatenação de strings para a mensagem normal
        print("Temperatura do sistema: " + str(temperature) + "°C. Tudo normal.")
except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro para a temperatura.")
```

```python
#MISSÃO 2
import os
import random
import shutil

# 1. Definir o diretório base para a bagunça e a organização
base_dir = './arquivos_baguncados'
organized_dir = './arquivos_organizados'

# Limpar diretórios anteriores para garantir um teste limpo
if os.path.exists(base_dir):
    shutil.rmtree(base_dir)
if os.path.exists(organized_dir):
    shutil.rmtree(organized_dir)

os.makedirs(base_dir, exist_ok=True)
os.makedirs(organized_dir, exist_ok=True)

print(f"Diretório de bagunça criado em: {base_dir}")
print(f"Diretório de organização criado em: {organized_dir}")

# 2. Criar 15 arquivos para teste
extensions = ['.txt', '.pdf', '.jpg', '.png', '.docx', '.xlsx', '.mp3']
file_names = []

print("\nCriando 15 arquivos de teste...")
for i in range(1, 16):
    ext = random.choice(extensions)
    file_name = f"documento_{i}{ext}"
    file_path = os.path.join(base_dir, file_name)
    with open(file_path, 'w') as f:
        f.write(f"Conteúdo do arquivo {file_name}")
    file_names.append(file_name)
    print(f"Criado: {file_name}")

print("\nArquivos de teste criados com sucesso.")

# 3. Organizar os arquivos
print("\nOrganizando arquivos...")
for filename in os.listdir(base_dir):
    if os.path.isfile(os.path.join(base_dir, filename)):
        file_name, file_extension = os.path.splitext(filename)

        # Remover o ponto da extensão para usar como nome da pasta
        folder_name = file_extension[1:].upper() if file_extension else 'SEM_EXTENSAO'
        destination_folder = os.path.join(organized_dir, folder_name)

        # Criar a pasta de destino se não existir
        os.makedirs(destination_folder, exist_ok=True)

        source_path = os.path.join(base_dir, filename)
        destination_path = os.path.join(destination_folder, filename)

        try:
            shutil.move(source_path, destination_path)
            print(f"Movido: {filename} para {folder_name}/")
        except Exception as e:
            print(f"Erro ao mover {filename}: {e}")

print("\nOrganização de arquivos concluída!")
print("Você pode verificar os resultados nos diretórios 'arquivos_baguncados' (agora vazio) e 'arquivos_organizados'.")
```

```python
#Missão 3
'''
Crie um programa autonomo que dispare alertas
(mensagens de alerta no console ou e-mail simulado)
unicamente quando uma condição critica for atingida.
'''
import time
import random

def check_critical_condition(current_value):
    """Simulates checking a critical condition (e.g., temperature)."""
    critical_threshold = 30 # Example: temperature > 30°C is critical
    return current_value > critical_threshold

def send_alert(message, alert_type='console', recipient='admin@example.com'):
    """Dispatches an alert based on the specified type."""
    if alert_type == 'console':
        print(f"\nALERTA CRÍTICO (Console): {message}")
    elif alert_type == 'email':
        print(f"\nALERTA CRÍTICO (E-mail Simulado):")
        print(f"De: sistema@autonomo.com")
        print(f"Para: {recipient}")
        print(f"Assunto: URGENTE - Condição Crítica Detectada!")
        print(f"Corpo: {message}\n")
    else:
        print(f"Tipo de alerta desconhecido: {alert_type}. Mensagem: {message}")

def autonomous_monitor(interval_seconds=5, max_iterations=5):
    """Simulates an autonomous monitoring system."""
    print("\n--- Iniciando Monitoramento Autônomo ---")
    print(f"Verificando a cada {interval_seconds} segundos. Total de {max_iterations} iterações.\n")

    iteration = 0
    while iteration < max_iterations:
        iteration += 1
        print(f"Iteração {iteration}/{max_iterations}:")

        # Simulate a sensor reading (e.g., temperature between 20 and 40)
        simulated_value = random.randint(20, 40)
        print(f"  Valor simulado do sistema: {simulated_value}°C")

        if check_critical_condition(simulated_value):
            alert_message = f"Temperatura atual de {simulated_value}°C excede o limite crítico de 30°C!"
            send_alert(alert_message, alert_type='console')
            send_alert(alert_message, alert_type='email', recipient='suporte@empresa.com.br')
        else:
            print(f"  Valor dentro dos limites aceitáveis. ({simulated_value}°C <= 30°C)")

        if iteration < max_iterations:
            print(f"  Aguardando {interval_seconds} segundos para a próxima verificação...")
            time.sleep(interval_seconds)

    print("\n--- Monitoramento Autônomo Finalizado ---")

# Para executar o programa autônomo, chame a função:
autonomous_monitor(interval_seconds=3, max_iterations=7)
```
